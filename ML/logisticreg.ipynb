```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets.samples_generator import make_blobs

def train_logistic_reg(X_train,y_test):
    w=[0,0,0]
    for i in range(1000):
        for j in range(len(X_train)):
            m=(w[0]+w[1]*X_train[j][0]+w[2]*X_train[j][1])
            s=1/(1+2.718**(-m))
            w[0]+=0.01*(y_train[j]-s)
            w[1]+=0.01*X_train[j][0]*(y_train[j]-s)
            w[2]+=0.01*X_train[j][1]*(y_train[j]-s)
    return w

def predict(X_test):
    y=[]
    for i in range(len(X_test)):        
        m=w[0]+w[1]*X_test[i][0]+w[2]*X_test[i][1]
        if m<0:
            y.append(0)
        else:
            y.append(1)
    return np.array(y)


w=train_logistic_reg(X_train,y_test)
m=-w[1]/w[2]
c=-w[0]/w[2]
predictions=predict(X_test)
accuracy=accuracy_score(y_test,predictions)
print('Accuracy:',accuracy)
plt.plot([1,10],[1*m+c,10*m+c],color="red")      

```
